### Create Scatterplots for each Feature by Outcome

```python
# Add a helper col to tennis data frame, which we will use to add a label for each row 
tennis['helper_col'] = 0

# Get the outcome columns
outcomes = ['helper_col', 'Wins', 'Losses', 'Winnings', 'Ranking']

# Get the feature columns
features = [name for name in tennis.columns if not (name in outcomes or name == "Player" or name == "Year")]

# Create the scatter plots
fig, axes = plt.subplots(len(features), len(outcomes), figsize=(30, 80))

# Loop through each feature first to get same features in same row, then outcomes to have same outcome in same column
for i, feature in enumerate(features):
    for j, outcome in enumerate(outcomes):
        axes[i, j].scatter(tennis[feature], tennis[outcome])
        axes[i, j].set_ylabel(outcome)
        axes[i, j].set_xlabel(feature)

for ax, col in zip(axes[0], outcomes):
    ax.set_title(col, fontweight='bold')

# Remove the first column of plots (scatter plots) from out helper col
for ax_row in axes:
    ax_row[0].remove()

# Add common feature labels on the left of each row (by labeling our removed helper col)
for i, feature in enumerate(features):
    axes[i, 0] = fig.add_subplot(len(features), len(outcomes) + 1, i * (len(outcomes) + 1) + 1)
    axes[i, 0].text(1.2, 0.5, feature, va='center', ha='center', fontsize=12, fontweight='bold', rotation='vertical')
    axes[i, 0].axis('off')

# Set the spacing between subplots
plt.subplots_adjust(left=0.15,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.2,
                    hspace=0.3)
plt.show()
```

**Summary:**<br>
1) Linear relationships of BreakPointsFaced, BreakPointsOpportunities, DoubleFaults, ReturnGamesWon, ServiceGamesWon with outcomes Wins, Losses, and Winnings. Here is what it means:<br>

#### BreakPointsFaced vs. Outcomes:
- **Wins**: A positive linear relationship between BreakPointsFaced and Wins might indicate that players who face more break points tend to win more matches, possibly indicating their ability to save crucial points.
- **Losses**: A positive linear relationship between BreakPointsFaced and Losses could suggest that facing more break points increases the likelihood of losing matches.
- **Winnings**: A negative linear relationship between BreakPointsFaced and Winnings could imply that players who frequently face break points might not perform as well in terms of earnings due to higher pressure situations.

#### BreakPointsOpportunities vs. Outcomes:
- **Wins**: A positive linear relationship between BreakPointsOpportunities and Wins might indicate that players who create more break point opportunities have a higher chance of winning matches.
- **Losses**: A positive linear relationship between BreakPointsOpportunities and Losses could suggest that despite creating opportunities, players are unable to capitalize on them, leading to losses.
- **Winnings**: A positive linear relationship between BreakPointsOpportunities and Winnings might imply that players who frequently create break point opportunities perform well financially due to their competitive play.

#### DoubleFaults vs. Outcomes:

- **Losses**: A positive linear relationship between DoubleFaults and Losses could indicate that riskier serving strategies might lead to more errors, resulting in losses.
- **Winnings**: A negative linear relationship between DoubleFaults and Winnings might imply that players with fewer double faults tend to perform better financially due to a more consistent serve.

## Linear Regression Model

```python
# Create a Linear Regression model
lm = LinearRegression()
```

```python
# Choose the predictor (feature) and target variables
X = tennis[['BreakPointsConverted']]  # Predictor variable(s)
y = tennis['Wins']                     # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Fit the model to the training data
model = lm.fit(X_train, y_train)

# Predict the target variable on the test set
y_predict = lm.predict(X_test)

# Print the R-squared scores for training and testing
print("R-square using train data:")
print(round(lm.score(X_train, y_train), 3))
print("R-square using test data:")
print(round(lm.score(X_test, y_test), 3))
```

**Summary:**<br>
The R-squared value of 0.008 when using the training data indicates that only a very small proportion of the variability in the target variable ('Wins' in this case) is explained by the predictor variable ('BreakPointsConverted' in this case) in the model. This suggests that the model doesn't fit the training data well and that the predictor variable 'BreakPointsConverted' might not have strong predictive power for the target variable 'Wins'. The low R-squared value indicates that there might be other factors not captured by the current predictor variable that influence the outcome variable.

## Multiple Linear Regression Model

```python
# Select the features and target variable
features = ['FirstServe', 'Aces']  # Features used for prediction
X = tennis[features]                # Feature variables
y = tennis['Wins']                  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Fit the model to the training data
model = lm.fit(X_train, y_train)

# Predict the target variable on the test set
y_predict = lm.predict(X_test)

# Print R-squared scores for training and testing
print("R-square using train data:")
print(round(lm.score(X_train, y_train), 3))
print("R-square using test data:")
print(round(lm.score(X_test, y_test), 3))
```

**Summary:**<br>
* The R-squared value of 0.692 obtained from the training data indicates that approximately 69.2% of the variability in the dependent variable ('Wins') can be explained by the linear regression model using the selected features 'First Serve' and 'Aces'. This suggests a relatively good fit of the model to the training data.<br>
* Similarly, the R-squared value of 0.649 obtained from the test data indicates that approximately 64.9% of the variability in the dependent variable can be explained by the model when applied to new, unseen data. This suggests that the model's performance on the test data is reasonably consistent with its performance on the training data, although it might slightly underperform compared to the training data.

```python
# Select features and target variable
features = ['Aces', 'DoubleFaults', 'FirstServe', 'FirstServePointsWon', 'SecondServePointsWon',
            'BreakPointsFaced', 'BreakPointsSaved', 'ServiceGamesPlayed', 'ServiceGamesWon', 'TotalServicePointsWon']
x = tennis[features]
y = tennis['Wins']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y)

# Fit the model to the training data
model = lm.fit(X_train, y_train)

# Predict the target variable on the test set
y_predict = lm.predict(X_test)

# Print R-squared scores for training and testing
print("R-square using train data:")
print(round(lm.score(X_train, y_train), 3))
print("R-square using test data:")
print(round(lm.score(X_test, y_test), 3))

# Create a scatter plot of predicted wins vs. actual wins
plt.scatter(y_test, y_predict, alpha=0.6)
plt.title('Predicted Wins vs. Actual Wins')
plt.xlabel('Actual Wins')
plt.ylabel('Predicted Wins')

# Add a regression line
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', linewidth=2, alpha=0.7)
plt.show()
```

**Summary:** <br>  
+ A high R-squared value of 0.894 on the training data indicates that approximately 89.4% of the variability in the 'Wins' target variable can be explained by the selected features in the model. This suggests a strong predictive relationship between the chosen features and the number of wins in the training dataset.

+ On the other hand, the R-squared value of 0.799 on the testing data indicates that the model performs well in explaining around 79.9% of the variability in 'Wins' for unseen data, reinforcing the model's generalization ability.

+ The scatter plot with a red dashed regression line also visualizes the model's predictions against the actual 'Wins' values, showcasing a positive correlation between the predicted and actual outcomes.

+ Overall, the model demonstrates a reasonably good fit to the data and shows promise in predicting tennis match outcomes based on the selected features.
